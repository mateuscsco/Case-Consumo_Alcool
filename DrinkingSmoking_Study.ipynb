<p style='font-size: 20px; line-height: 2'><b>Explicando as métricas acima:<p style='font-size: 20px; line-height: 2'><b>
    
- **Accuracy (Acurácia):**
  - A proporção de exemplos classificados corretamente em relação ao total de exemplos.
  - Fórmula: `(Verdadeiros Positivos + Verdadeiros Negativos) / Total de Exemplos`
  - Indica o quão bem o modelo está performando globalmente.

- **Precision (Precisão):**
  - A proporção de verdadeiros positivos em relação ao total de exemplos previstos como positivos (soma de verdadeiros positivos e falsos positivos).
  - Fórmula: `Verdadeiros Positivos / (Verdadeiros Positivos + Falsos Positivos)`
  - Indica a precisão do modelo quando prevê positivos.

- **Recall (Revocação ou Sensibilidade):**
  - A proporção de verdadeiros positivos em relação ao total de exemplos verdadeiramente positivos (soma de verdadeiros positivos e falsos negativos).
  - Fórmula: `Verdadeiros Positivos / (Verdadeiros Positivos + Falsos Negativos)`
  - Indica a capacidade do modelo de encontrar todos os exemplos positivos.

- **F1-Score:**
  - A média harmônica entre precision e recall. É uma métrica que busca um equilíbrio entre precisão e sensibilidade.
  - Fórmula: `2 * (Precision * Recall) / (Precision + Recall)`
  - Oferece uma visão única que considera ambos os falsos positivos e falsos negativos.

- **Macro Avg (Média Macro):**
  - Calcula a média das métricas para cada classe sem levar em consideração o desbalanceamento de classes.
  - Calcula a média aritmética das métricas de cada classe.

- **Weighted Avg (Média Ponderada):**
  - Calcula a média ponderada das métricas para cada classe, levando em consideração o número de instâncias de cada classe.
  - Útil quando há um desbalanceamento significativo entre as classes.

- **Support:**
  - O número total de instâncias de cada classe no conjunto de teste.
  - Indica a quantidade de dadosasses são relativamente equilibrados nos três modelos.</b></p>
